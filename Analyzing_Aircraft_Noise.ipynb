```python
#importing liberaries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
NE_df = pd.read_csv('Noise_Exceedance_Rating.csv')
```

```python
Airline_names = NE_df['Airline'].unique()
Airline_names
```

```python
len(Airline_names)
```

```python
df_2006 = NE_df[NE_df['Year'] == 2006]
df_2006
```

```python
df_2006 = df_2006[df_2006['Month'] == 'January']
df_2006
```

```python
df_UAL = NE_df[NE_df['Airline'] == 'UNITED AIRLINES']
```

```python
df_UAL
```

```python
df_SIA = NE_df[NE_df['Airline'] == 'SINGAPORE AIRLINES']
df_SIA 
```

```python
df_ACA = NE_df[NE_df['Airline'] == 'AIR CANADA']
df_ACA
```

```python
# group the data by year
grouped = df_UAL.groupby(df_UAL['Year'])

# create a stacked bar graph
grouped[['Total Noise Exceedances', 'Total Operations per Month']].sum().plot(kind='bar', stacked=True)

# set the x-axis label
plt.xlabel('Year')

# set the y-axis label
plt.ylabel('Number of Operations')

# set the title
plt.title('Total Noise Exceedances and Total Operations per Month over the Years "United Airline"')

# show the plot
plt.show()
```

# Reading new CSV for Cleaning

```python
NC_df = pd.read_csv('Aircraft_Noise_Complaint_Data.csv')
NC_df
```

```python
# Create a dictionary of month numbers and names
month_names = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}

# Map the month numbers to their names in the DataFrame
NC_df['Month'] = NC_df['Month'].map(month_names)
NC_df
```

```python
unique_names = NC_df['Community'].unique()

print(unique_names)
```

```python
NC_df = NC_df.rename(columns={'Community': 'City'})
NC_df
```

```python
# Export the DataFrame to a CSV file
NC_df.to_csv('Noise_Complaints_around_globe.csv', index=False)
```
